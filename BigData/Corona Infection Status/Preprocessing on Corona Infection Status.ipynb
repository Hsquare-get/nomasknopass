# Load Raw Data on Corona Infection Status

```python
import pandas as pd

corona_df = pd.read_csv('./corona_stat(0215~1208).csv')

print(corona_df.info()) # corona_df.dtypes
display(corona_df)
```

# Check for Missing Data

```python
print(corona_df.isnull().sum())
```

# Check for Duplicated Data

### seq열에서 중복행 제거(처음 나타난 데이터는 보존 = updatedt가 있는 행)

```python
# seq 열의 고유값 개수 카운트(NaN 포함)
seq_unique = corona_df['seq'].value_counts(dropna=False) # Series
seq_unique = pd.DataFrame(seq_unique) # DataFrame
display(seq_unique)

# seq 열에서 중복된 값이 있는 행의 수
counterFunc = seq_unique.apply(lambda x: True if x[0] >= 2 else False, axis=1)
numOfRows = len(counterFunc[counterFunc == True].index)
print('게시글 번호(seq)가 중복된 행의 수:', numOfRows)

display(corona_df.loc[(corona_df['seq'] == 289) | (corona_df['seq'] == 338) | (corona_df['seq'] == 277)])
```

```python
corona_df = corona_df.drop_duplicates(subset=['seq']) # seq열에서 중복행 제거(처음 나타난 데이터는 보존 = updatedt가 있는 행)
display(corona_df)
```

### statedt열에서 중복행 제거

```python
# statedt 열의 고유값 개수 카운트(NaN 포함)
statedt_unique = corona_df['statedt'].value_counts(dropna=False) # Series
statedt_unique = pd.DataFrame(statedt_unique) # DataFrame
display(statedt_unique.loc[statedt_unique['statedt'] >= 2])

# statedt 열에서 중복된 값이 있는 행의 수
counterFunc = statedt_unique.apply(lambda x: True if x[0] >= 2 else False, axis=1)
numOfRows = len(counterFunc[counterFunc == True].index)
print('기준일(statedt)이 중복된 행의 수:', numOfRows)

display(corona_df.loc[corona_df['statedt']==20200425])
```

# Convert to Datetime

```python
import datetime as dt


def strptime_statedt(x):
    x = str(x)
    return dt.datetime.strptime(x, '%Y%m%d')

def strptime_statetime(x):
    x = str(x)
    return dt.datetime.strptime(x, '%H:%M').hour

corona_df['statedt'] = corona_df['statedt'].apply(strptime_statedt)
corona_df['statetime'] = corona_df['statetime'].apply(strptime_statetime)
corona_df['createdt'] = pd.to_datetime(corona_df['createdt'])
corona_df['updatedt'] = pd.to_datetime(corona_df['updatedt'])

print(corona_df.info())
display(corona_df)
```

### 4월 25일 이상치 발견, statetime 기준 내림차순이 아닌 createtime 기준 내림차순으로 변경

```python
# statedt(기준일) 기준으로는 오름차순, createdt(등록일시분초) 기준으로는 내림차순 정렬
corona_df = corona_df.sort_values(by=["statedt", "createdt"], ascending=[True, False])
corona_df = corona_df.drop_duplicates(subset=['statedt']) # statedt열에서 중복행 제거(처음 나타난 데이터는 보존 = 그 날의 가장최신 데이터)

display(corona_df)
```

### carecnt가 없는 날은 -1처리 해준것을 0으로 처리 (clearcnt는 없는 날이 없었음)

```python
print(corona_df['carecnt'].min()) # -1
print(corona_df['clearcnt'].min()) # 9

corona_df['carecnt'].replace(-1, 0, inplace=True)
display(corona_df)
```

```python
def str_slicing(x):
    x = str(x)
    x = x[:10]
    return x

corona_df['statedt'] = corona_df['statedt'].apply(str_slicing)
print(corona_df.info())
```

# Save the Preprocessed Data for Corona Infection Status
