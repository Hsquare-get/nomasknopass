## Search periods (2020.02.15 ~ 2020.12.02)

```python
import urllib.request
from bs4 import BeautifulSoup # HTML, XML Document 데이터를 읽기 위한 라이브러리
import xml.dom.minidom

serviceKey = '0Nxjl3xb59fkPcq7n0%2FoeaP1%2BvZy8L1l%2B4NunrGvuPrFcORrGfI2caBJ2dzsJA6VZWYw6XGIJ0hH2%2Bx9V74bUQ%3D%3D'
startCreateDt = '20200215'
endCreateDt = '20201202'
url_xml = str('http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?serviceKey=' + serviceKey + 
              '&startCreateDt=' + startCreateDt + '&endCreateDt=' + endCreateDt)

response = urllib.request.urlopen(url_xml)
response_code = response.getcode()

if response_code == 200:
    response_body = response.read().decode('utf-8') # decode('utf-8') : 바이트열 -> 문자열(복호화)
    soup = BeautifulSoup(response_body, 'html.parser') # xml도 가능

    totalcount = soup.find('totalcount').text
    dataList = soup.find_all('item')
    
    infection_status = {'seq':[] ,'statedt':[], 'statetime':[], 'createdt':[], 'updatedt':[], 
                        'decidecnt':[], 'examcnt':[], 'deathcnt':[], 'carecnt':[], 'clearcnt':[]}
    for data in dataList:
        seq = data.seq.text
        infection_status['seq'].append(seq)
        statedt = data.statedt.text
        infection_status['statedt'].append(statedt)
        statetime = data.statetime.text
        infection_status['statetime'].append(statetime)
        createdt = data.createdt.text
        infection_status['createdt'].append(createdt)
        updatedt = data.updatedt.text
        infection_status['updatedt'].append(updatedt)
        
        decidecnt = data.decidecnt.text
        infection_status['decidecnt'].append(decidecnt)
        examcnt = data.examcnt.text
        infection_status['examcnt'].append(examcnt)
        deathcnt = data.deathcnt.text
        infection_status['deathcnt'].append(deathcnt)
        carecnt = data.carecnt.text if data.carecnt else -1
        infection_status['carecnt'].append(carecnt)
        clearcnt = data.clearcnt.text if data.clearcnt else -1
        infection_status['clearcnt'].append(clearcnt)
        
    # xml문서를 최신부터 읽기 때문에 시간순으로 데이터를 집어넣으려면 reverse
    infection_status['seq'].reverse()
    infection_status['statedt'].reverse()
    infection_status['statetime'].reverse()
    infection_status['createdt'].reverse()
    infection_status['updatedt'].reverse()
    
    infection_status['decidecnt'].reverse()
    infection_status['examcnt'].reverse()
    infection_status['deathcnt'].reverse()
    infection_status['carecnt'].reverse()
    infection_status['clearcnt'].reverse()
    
    
print('days:', totalcount)
print(infection_status['decidecnt'][-1], '/ dataNum:', len(infection_status['decidecnt']))
print(infection_status['examcnt'][-1], '/ dataNum:', len(infection_status['examcnt']))
print(infection_status['deathcnt'][-1], '/ dataNum:', len(infection_status['deathcnt']))
```
