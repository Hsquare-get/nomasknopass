# Mask plot 및 좌표입력 (마스크 추가할 때 좌표값 알아내는 용도)

```python
import cv2
from matplotlib import pyplot as plt
```

```python
red_color = (255, 0, 0)

a = cv2.imread('masks/templates/surgical_left.png')

a_1 = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)

a_2 = cv2.line(a_1, (39,27),(39,27),red_color,5)   # 상단 왼쪽
a_2 = cv2.line(a_1, (130,9),(130,9),red_color,5)   # 상단 중앙
a_2 = cv2.line(a_1, (567,20),(567,20),red_color,5)   # 상단 오른쪽
a_2 = cv2.line(a_1, (30,207),(30,207), red_color,5)   # 하단 왼쪽
a_2 = cv2.line(a_1, (120,302),(120,302),red_color,5)   # 하단 중앙 
a_2 = cv2.line(a_1, (568,202),(568,202),red_color,5)    # 하단 오른쪽
plt.imshow(a_2)

## masks.cfg 파일에 좌표 값이랑 마스크 파일 위치 등록해줘야함. 
```
